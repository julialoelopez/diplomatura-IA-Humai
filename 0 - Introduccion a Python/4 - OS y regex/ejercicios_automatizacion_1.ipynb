```python
import requests
import re
import os
```

```python
# Nos traemos un peque√±o dataset (con info falsa) tomado de este excelente video: https://youtu.be/K8L6KVGG-7o
# Hacemos un request (spoiler alert: vamos a ver m√°s de esto la pr√≥xima clase üòâ)
r = requests.get('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python-Regular-Expressions/data.txt')
texto = r.text
print(texto)
```

```python
# Patr√≥n regex para encontrar URLs
pattern = re.compile(r'[A-Z]{2}\s\d{5}')

# Encuentra todos los URLs en la cadena
matches = pattern.findall(texto)

# Imprime los URLs encontrados
for match in matches:
    print(match)
```

## Ejercicios

1. Crear una carpeta *datos*.
2. Dentro de esta nueva carpeta, crear diferentes carpetas para nombres, telefono , emails, y c√≥digo de estado extraidos del texto.
3. Seleccionar solo los numeros de telefono que empiecen con 2 o con 3.
4. Guardar en formato txt un archivo en cada carpeta con la data correspondiente <br>
(**TIP: Agreguen saltos de linea "\n" luego de cada entrada**).



```python
if not os.path.exists('datos'):
    os.mkdir('datos')
    os.chdir('datos')
carpetas = ['nombres', 'telefonos', 'emails', 'codigos de estado']
for carpeta in carpetas:
    if not os.path.exists(f'{carpeta}'):
        os.mkdir(f'{carpeta}')
```

```python
regla_de_busqueda = "[2-3]\d{2}-\d{3}-\d{4}"
re.findall(regla_de_busqueda, texto)
```

```python
os.chdir('..')
os.getcwd()
```

```python
# crear un archivo para cada tipo de dato y guardarlo en la carpeta correspondiente
#Primero una expresion regular para encontrar todos los nombres, otra pars telefonos, otra para emails y otra para codigos de estado
#Luego un for para cada una de las expresiones regulares y guardar los resultados en un archivo de texto
#Cada archivo de texto debe estar en la carpeta correspondiente
regex_nombre = r'\n([A-Z]\w+\s[A-Z]\w+)'
nombres = re.findall(regex_nombre, texto)
regex_telefono = r'\d{3}-\d{3}-\d{4}'
telefonos = re.findall(regex_telefono, texto)
regex_email = r'[\w\.-]+@[\w\.-]+'
emails = re.findall(regex_email, texto)
regex_codigo = r'[A-Z]{2}\s\d{5}'
codigos = re.findall(regex_codigo, texto)

```

```python
if not os.path.exists('nombre.txt'):
    with open('nombre.txt', 'w') as f:
        for nombre in nombres:
            f.write(nombre + '\n')
if not os.path.exists('telefono.txt'):
    with open('telefono.txt', 'w') as f:
        for telefono in telefonos:
            f.write(telefono + '\n')
if not os.path.exists('email.txt'):
    with open('email.txt', 'w') as f:
        for email in emails:
            f.write(email + '\n')
if not os.path.exists('codigo.txt'):
    with open('codigo.txt', 'w') as f:
        for codigo in codigos:
            f.write(codigo + '\n')
            
```

```python
urls = '''
https://www.google.com
http://data.buenosaires.gob.ar
https://youtube.com
https://www.nasa.gov
https://humai.com.ar
https://www.7puentes.com
'''

Necesito que me generes un patr√≥n de regex que permita extraer contenido de urls como en los siguientes ejemplos:

https://www.google.com ->google
http://data.buenosaires.gob.ar ->data.buenosaires
https://youtube.com -> youtube
https://www.nasa.gov ->
https://humai.com.ar ->
https://www.7puentes.com ->
```

```python
import re

urls = '''
https://www.google.com
http://data.buenosaires.gob.ar
https://youtube.com
https://www.nasa.gov
https://humai.com.ar
https://www.7puentes.com
'''

# Patr√≥n regex para encontrar URLs
pattern = re.compile(r'https?://(?:www\.)?([^\.]+)\.[^\/]+')

# Encuentra todos los URLs en la cadena
matches = pattern.findall(urls)

# Imprime los URLs encontrados
for match in matches:
    print(match)
```

```python
import re

urls = '''
https://www.google.com
http://data.buenosaires.gob.ar
https://youtube.com
https://www.nasa.gov
https://humai.com.ar
https://www.7puentes.com
'''

# Patr√≥n regex para encontrar URLs
pattern = re.compile(r'https?://\S+')

# Encuentra todos los URLs en la cadena
matches = pattern.findall(urls)

# Imprime los URLs encontrados
for match in matches:
    print(match)
```
